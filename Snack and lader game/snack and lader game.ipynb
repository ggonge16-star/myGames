```python
import random

# Snake and ladder positions
snakes = {16: 6, 47: 26, 49: 11, 56: 53, 62: 19, 64: 60, 87: 24, 93: 73, 95: 75, 98: 78}
ladders = {1: 38, 4: 14, 9: 31, 21: 42, 28: 84, 36: 44, 51: 67, 71: 91, 80: 100}

def roll_dice():
    return random.randint(1, 6)

def move_player(player, pos):
    pos += roll_dice()
    if pos in snakes:
        print(f"Oh no! {player} bitten by snake! Go down to {snakes[pos]}")
        pos = snakes[pos]
    elif pos in ladders:
        print(f"Yay! {player} climbed ladder to {ladders[pos]}")
        pos = ladders[pos]
    if pos > 100:
        pos = 100
    return pos

def play_game():
    player_pos = 0
    computer_pos = 0

    while True:
        input("Press Enter to roll dice...")
        player_pos = move_player("You", player_pos)
        print(f"You are now at {player_pos}\n")

        if player_pos == 100:
            print("Congratulations! You won!")
            break

        print("Computer's turn...")
        computer_pos = move_player("Computer", computer_pos)
        print(f"Computer is now at {computer_pos}\n")

        if computer_pos == 100:
            print("Computer won! Better luck next time.")
            break

if __name__ == "__main__":
    print("Welcome to Snake and Ladder Game!")
    play_game()
```
